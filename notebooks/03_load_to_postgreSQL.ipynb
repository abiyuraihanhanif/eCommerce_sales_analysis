```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv(
    "../data/ecommerce_data_clean.csv"
)

df.head()
```

```python
username = "postgres"
password = "Abi_28102000"
host = "localhost"
port = "5432"
database = "ecommerce_sales"

engine = create_engine(
    f"postgresql://{username}:{password}@{host}:{port}/{database}"
)
```

```python
engine.connect()
```

```python
df.to_sql(
    "ecommerce_sales",
    engine,
    if_exists="replace",
    index=False
)
```

```python
#verivikasi data
query = """
SELECT COUNT(*)
FROM ecommerce_sales
"""

pd.read_sql(query, engine)
```

```python
#Cek SQL sudah bisa berjalan atau belum
query = """
SELECT
    category,
    SUM(sales) revenue
FROM ecommerce_sales
GROUP BY category
ORDER BY revenue DESC
"""

pd.read_sql(query, engine)
```
