```python
import pandas as pd

df = pd.read_csv("../data/ecommerce_sales_data.csv")

#Missing Value
df.isnull().sum()
```

```python
#Duplicate
df.duplicated().sum()
```

```python
#Data Type
df.dtypes
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Descriptive Statistic
df.describe().round(2)
```

```python
# Business Metrics Summary
print(f"Total Sales   : {df['Sales'].sum():,.2f}")
print(f"Total Profit  : {df['Profit'].sum():,.2f}")
print(f"Total Quantity: {df['Quantity'].sum():,.0f}")
```

```python
# Loss Transactions
loss_transactions = (df['Profit'] < 0).sum()

print(f"Jumlah transaksi rugi: {loss_transactions}")
print(f"Persentase transaksi rugi: {(loss_transactions/len(df))*100:.2f}%")
```

```python
#Profit Margin
profit_margin = (df['Profit'].sum() / df['Sales'].sum()) * 100

print(f"Profit Margin: {profit_margin:.2f}%")
```

```python
summary = pd.DataFrame({
    'Metric': ['Total Rows',
               'Total Sales',
               'Total Profit',
               'Total Quantity',
               'Unique Products',
               'Unique Categories',
               'Unique Regions'],
    'Value': [
        len(df),
        df['Sales'].sum(),
        df['Profit'].sum(),
        df['Quantity'].sum(),
        df['Product Name'].nunique(),
        df['Category'].nunique(),
        df['Region'].nunique()
    ]
})

summary
```
